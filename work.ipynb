```python
import pandas as pd
import numpy as np
import json

#Group_C 분리된 csv 파일로 작업함
train = pd.read_csv('train_C_Change.csv', encoding='euc-kr')

train
```

```python
#value만 뽑기
train_A = train['Feature_Group_A']
train_B = train['Feature_Group_B']
train_printabledist = train['printabledist']
```

```python
train_A_val = train_A.values
train_B_val = train_B.values
train_printabledist_val = train_printabledist.values
```

```python
#list화
train_A_list = train_A_val.tolist()
train_B_list = train_B_val.tolist()
train_printabledist_list = train_printabledist_val.tolist()
```

```python
## 따옴표 제거
A_list=[json.loads(x) for x in train_A_list]
B_list=[json.loads(x) for x in train_B_list]
printabledist_list=[json.loads(x) for x in train_printabledist_list]

len(printabledist_list[0])
```

```python
# DataFrame columns 만들기용 숫자 list
A_name_list = ["A%d" % i for i in range(1,257)]
B_name_list = ["B%d" % i for i in range(1,257)]
printabledist_name_list = ["printabledist%d" % i for i in range(1,len(printabledist_list[0])+1)]
```

```python
## A_Group DataFrame
feature_group_a_df = pd.DataFrame(columns = A_name_list)
    
## B_Group DataFrame
feature_group_b_df = pd.DataFrame(columns = B_name_list)
    
## printabledist DataFrame
feature_group_p_df = pd.DataFrame(columns = printabledist_name_list)


#insert data
for i in range(10000):
    feature_group_a_df = feature_group_a_df.append(pd.Series(A_list[i], index=feature_group_a_df.columns), ignore_index=True)
    feature_group_b_df = feature_group_b_df.append(pd.Series(B_list[i], index=feature_group_b_df.columns), ignore_index=True)
    feature_group_p_df = feature_group_p_df.append(pd.Series(printabledist_list[i], index=feature_group_p_df.columns), ignore_index=True)
```

```python
# Dataset 병합

train_change = pd.concat([train, feature_group_p_df], axis=1)
train_change = pd.concat([train_change, feature_group_a_df], axis=1)
train_change = pd.concat([train_change, feature_group_b_df], axis=1)
```

```python
train_change = train_change.drop(columns=['printabledist'], axis=1)
train_change = train_change.drop(columns=['Feature_Group_A'], axis=1)
train_change = train_change.drop(columns=['Feature_Group_B'], axis=1)
train_change = train_change.drop(columns=['Unnamed: 0'], axis=1)
```
